# SMA

```python
# Filter the date index and close columns
sma_signals_df = data_df.loc[:, ["Close"]]

# Use the pct_change function to generate returns from close prices
sma_signals_df["Actual Returns"] = sma_signals_df["Close"].pct_change()

# Drop all NaN values from the DataFrame
sma_signals_df = sma_signals_df.dropna()

# Review the DataFrame
display(sma_signals_df.head())
display(sma_signals_df.tail())
```

```python
#Set the short window and long window 120day
short_window = 30
long_window = 120

# Generate the fast and slow simple moving averages (4 and 100 days, respectively)
sma_signals_df['30_SMA'] = sma_signals_df['Close'].rolling(window=short_window).mean()
sma_signals_df['120_SMA'] = sma_signals_df['Close'].rolling(window=long_window).mean()

sma_signals_df = sma_signals_df.dropna()

# Review the DataFrame
display(sma_signals_df.head())
display(sma_signals_df.tail())
```

```python
# Initialize the new Signal column
sma_signals_df['Signal'] = 0.0

# When Actual Returns are greater than or equal to 0, generate signal to buy stock long
sma_signals_df.loc[(sma_signals_df['Actual Returns'] >= 0), 'Signal'] = 1

# When Actual Returns are less than 0, generate signal to sell stock short
sma_signals_df.loc[(sma_signals_df['Actual Returns'] < 0), 'Signal'] = -1

# Review the DataFrame
display(sma_signals_df.head())
display(sma_signals_df.tail())
```

```python
#View value count of the signals for sma
sma_signals_df['Signal'].value_counts()
```

```python
# Calculate the strategy returns and add them to the signals_df DataFrame
sma_signals_df['Strategy Returns'] = sma_signals_df['Actual Returns'] * sma_signals_df['Signal'].shift()

# Review the DataFrame
display(sma_signals_df.head())
display(sma_signals_df.tail())
```

```python
def calculate_force_index(data, ndays):
    # Calculate raw Force Index
    data['ForceIndex'] = data['Close'].diff(ndays) * data['Volume']

    # Standardize the Force Index
    scaler = StandardScaler()
    data['StandardizedForceIndex'] = scaler.fit_transform(data['ForceIndex'].values.reshape(-1, 1))

    return data

def prepare_data_for_classification(data, prediction_horizon=1):
    # Assuming you want to predict the price direction 1 day ahead (up or down)
    data['PriceDirection'] = (data['Close'].shift(prediction_horizon) > data['Close']).astype(int)
    
    # Drop rows with NaN values created by shifting
    data = data.dropna()

    # Features and target variable
    X = data[['StandardizedForceIndex']]
    y = data['PriceDirection']

    return X, y
data
```

```python
# Calculate RSI
def calculate_rsi(rsi_data_df, period=14):
    close_diff = rsi_data_df['Close'].diff(1)
    
    # Calculate gain and loss
    gain = close_diff.where(close_diff > 0, 0)
    loss = -close_diff.where(close_diff < 0, 0)

    # Calculate average gain and average loss
    avg_gain = gain.rolling(window=period, min_periods=1).mean()
    avg_loss = loss.rolling(window=period, min_periods=1).mean()

    # Calculate relative strength (RS)
    rs = avg_gain / avg_loss

    # Calculate RSI
    rsi = 100 - (100 / (1 + rs))

    rsi_data_df['RSI'] = rsi
    return rsi_data_df
```

```python
#retrieve data
rsi_data_df = calculate_rsi(rsi_data_df)

#preview data

rsi_data_df
```

```python
# Implement RSI strategy
def rsi_strategy(rsi_data_df, overbought_threshold=70, oversold_threshold=30):
    rsi_signals = pd.DataFrame(rsi_data_df)

    # Generate signals
    rsi_signals['buy_signal'] = 0.0
    rsi_signals['sell_signal'] = 0.0

    # Buy signal
    rsi_signals.loc[rsi_data_df['RSI'] < oversold_threshold, 'buy_signal'] = 1

    # Sell signal
    rsi_signals.loc[rsi_data_df['RSI'] > overbought_threshold, 'sell_signal'] = -1

    return rsi_signals

rsi_signals = rsi_strategy(rsi_data_df)

#preview
rsi_signals
```
